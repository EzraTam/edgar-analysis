```python
def extract_yearly_data_from_df(eps):
    df_res=pd.DataFrame(eps)
    df_res = df_res[df_res["fp"] == "FY"]
    df_res["end"] = pd.to_datetime(df_res["end"], format="%Y-%m-%d")
    df_res["year"] = pd.DatetimeIndex(df_res["end"]).year
    return (
        df_res.sort_values(["year", "fy"], ascending=False)
        .groupby("year")
        .head(1)[["year", "val"]]
        .reset_index(drop=True)
        )
```

```python
df_eps=pd.DataFrame(columns=["cik","company","year","eps"])
for file_nm in li_comp:
    with open(os.path.join(PATH_DATA, file_nm), encoding="utf-8") as file:
        # Define service
        raw_data=json.load(file)
        try:
            cik=raw_data["cik"]
        except:
            cik=None
        try:
            entity_name=raw_data["entityName"]
        except:
            entity_name=None
        try:
            eps=raw_data["facts"]["us-gaap"]["EarningsPerShareDiluted"]["units"]["USD/shares"]
        except:
            eps=None
        if eps:
            _df_temp=extract_yearly_data_from_df(eps)
            _df_temp["cik"]=cik
            _df_temp["company"]=entity_name
            df_eps=pd.concat([df_eps,_df_temp],ignore_index=True)
```
